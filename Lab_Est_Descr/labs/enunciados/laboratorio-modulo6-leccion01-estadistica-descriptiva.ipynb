#### Ejercicio 1: 

1.  Dados los siguientes datos: [15, 21, 25, 32, 35, 45, 50, 50, 55, 60]. Crea una funcion que calcule la media, mediana y moda.


```python
lista1 = [15, 21, 25, 32, 35, 45, 50, 50, 55, 60]
```

```python
def convierte_df(lista):
    df = pd.DataFrame(lista)
    return df
```

```python
df = convierte_df(lista1)
```

```python
df.mean()
```

```python
def est_descriptivas(df):
    media = df[0].mean()
    mediana = df[0].median()
    moda = df[0].mode()[0]
    return (f'El resultado de mis estadísticas es una media de {media}, una mediana de {mediana} y una moda de {moda}')
```

```python
est_descriptivas(df)
```
