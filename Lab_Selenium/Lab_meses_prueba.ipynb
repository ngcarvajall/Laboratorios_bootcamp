```python
# para empezar a extraer
def llamada_tabla(sopa,municipio):
       lista_tablas = sopa.find_all('table')
       print(type(lista_tablas))
       columnas = lista_tablas[1]
       columnas= [l.getText() for l in columnas.findAll('thead')]
       time = lista_tablas[2]
       temperature = lista_tablas[3]
       dewpoint = lista_tablas[4]
       humidity = lista_tablas[5]
       wspeed= lista_tablas[6]
       pressure = lista_tablas[7]
       precipitation = lista_tablas[8]

       tabla = [tiempo.getText() for tiempo in time.findAll('td')]
       tabla_1= [t.getText() for t in temperature.findAll('td')]
       tabla_2= [d.getText() for d in dewpoint.findAll('td')]
       tabla_3= [h.getText() for h in humidity.findAll('td')]
       tabla_4= [ws.getText() for ws in wspeed.findAll('td')]
       tabla_5= [ps.getText() for ps in pressure.findAll('td')]
       tabla_6= [precip.getText() for precip in precipitation('td')]

       nombre =[tabla[0].strip()]
       datos = tabla[1:]
       mes=tabla[0]
       dfoct = pd.DataFrame(datos, columns=nombre)



       column_1 = tabla_1[:3]

       data_1 = tabla_1[3:]

       reshaped_data = np.array(data_1).reshape(-1, 3)

       df1 = pd.DataFrame(reshaped_data, columns=column_1)

       column_2 = tabla_2[:3]

       data_2 = tabla_2[3:]

       reshaped_data2 = np.array(data_2).reshape(-1, 3)

       df2 = pd.DataFrame(reshaped_data2, columns=column_2)

       column_3 = tabla_3[:3]

       data_3 = tabla_3[3:]

       reshaped_data3 = np.array(data_3).reshape(-1, 3)

       df3 = pd.DataFrame(reshaped_data3, columns=column_3)

       column_4 = tabla_4[:3]

       data_4 = tabla_4[3:]

       reshaped_data4 = np.array(data_4).reshape(-1, 3)

       df4 = pd.DataFrame(reshaped_data4, columns=column_4)

       column_5 = tabla_5[:3]

       data_5 = tabla_5[3:]

       reshaped_data5 = np.array(data_5).reshape(-1, 3)

       df5 = pd.DataFrame(reshaped_data5, columns=column_5)

       name = [tabla_6[0].strip()]
       values = tabla_6[1:]
       df7= pd.DataFrame(values, columns=name)

       df_final = pd.concat([dfoct,df1, df2, df3, df4, df5,df7 ], axis=1)

       df_final.columns = ['Día', 'Temp Max ', 'Temp Avg ', 'Temp Min ', 'Dew Point Max ', 'Dew Point Avg ', 'Dew Point Min ', 'Humidity (%) Max ',
              'Humidity (%) Avg ', 'Humidity (%) Min ', 'Wind Speed (mph) Max ', 'Wind Speed (mph) Avg ', 'Wind Speed (mph) Min ', 'Pressure (in) Max ', 'Pressure (in) Avg ', 'Pressure (in) Min ',
              'Precipitation Total']

       # diccionario = {1: 'Enero', 2: 'Febrero', 3: 'Marzo', 4: 'Abril', 5: 'Mayo', 6: 'Junio', 7:'Julio', 8: 'Agosto', 9: 'Septiembre', 10: 'Octubre', 11:'Noviembre', 12: 'Diciembre'}
       df_final.insert(column="Mes",loc = 0, value=mes)
       df_final.insert(column='Municipio', loc= 0, value=municipio)
       return df_final
```
