#### Ejercicio 11: ¿Cuál es total de tipos de calificaciones, considerando solo los principales 5 países? Deberás:

- Poner nombre a la ejes. 

- Poner título a la gráfica. 

- Rota las etiquetas del eje X. 

```python
# Filtrar los principales 5 países por cantidad de calificaciones
df_agrupado = netflix_mergeo.groupby("country")[["rating"]].count().nlargest(5, "rating").reset_index()

# Filtrar el dataframe original para solo los principales 5 países
top_5_paises = df_agrupado["country"]
df_filtrado = netflix_mergeo[netflix_mergeo["country"].isin(top_5_paises)]

plt.figure(figsize=(12,6))
sns.countplot(x="country", 
              hue="rating", 
              data=df_filtrado, 
              palette="viridis")
plt.title("Distribución de tipos de calificaciones en los 5 principales países")
plt.xlabel("Países")
plt.ylabel("Cantidad total de calificaciones")
plt.xticks(rotation=45)
plt.legend(title="Tipo de calificación")
plt.show()
```

#### Ejercicio 12: ¿Cómo se distribuyen las 5 primeras calificaciones  a lo largo de los aúltimo 10 años? 

- Poner nombre a la ejes. 

- Poner título a la gráfica. 

- Rota las etiquetas del eje X. 

- Elimina el nombre del eje X. 

```python
principales_ratings = ['TV-MA', 'TV-14', 'TV-PG', 'R', 'PG-13']
filtro_rating = netflix_mergeo["rating"].isin(principales_ratings)

filtro_yr = netflix_mergeo["release_year"].between(netflix_mergeo["release_year"].max()-10,netflix_mergeo["release_year"].max(),inclusive = "both")

df_filtrado = netflix_mergeo[filtro_rating & filtro_yr].groupby(["rating","release_year"])[["title"]].count().reset_index()

plt.figure(figsize=(10,5))
sns.lineplot(x="release_year",
             y="title",
             hue='rating',
             data=df_filtrado,
             linewidth = 1,
             linestyle = "dotted", 
             marker = "o");
plt.title('Distribución de los 5 principales ratings en los últimos 10 años')
plt.xticks(rotation=45)
plt.xlabel('')
plt.ylabel('Cantidad de películas')
plt.show()
```
