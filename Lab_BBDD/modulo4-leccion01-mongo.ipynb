# ¿Qué es Mongo DB?

Una base de datos no relacional, también conocida como NoSQL (not only SQL), no almacena la información en filas, lo que permite almacenar esta información de forma más flexible y menos estructurada.

MongoDB es una base de datos no relacional documental, es decir, que almacena documentos, cuyos registros se almacenan en archivos BSON (derivados de los archivos tipo JSON).

Definamos ahora algunos conceptos básicos que nos vamos a encontrar en MongoDB.
* **Campo**: Único dato almacenado dentro de un documento. En una base de datos relacional sería un dato individual de una columna.

* **Documentos**: Serían los registros de nuestra colección. Sería el equivalente a las filas de una base de datos SQL

* **Colecciones**: Son agrupaciones de documentos. Si hiciéramos una analogía con las bases de datos relacionales serían equivalentes a las tablas, salvo que en este caso no tienen un esquema fijo para guardar la información.

* **Bases de datos**: Son el conjunto de nuestras colecciones. Para una base de datos SQL sería el conjunto de sus tablas con las relaciones entre ellas.


Además, MongoDB acepta unos tipos de datos concretos. Los principales los vamos a enumerar a continuacion, el resto puedes encontrarlos en la documentación.

* **String**: Representa cadenas de texto.

* **Number**: Representa números enteros o decimales.

* **Boolean**: Representa valores booleanos true/false (True/False)
.
* **Date**: Representa fechas y horas.

* **ObjectId**: Representa identificadores únicos de documentos en MongoDB.

* **Array**: Representa una lista ordenada de elementos.

* **Embedded** Document: Representa documentos embebidos dentro de otros documentos.

* **Binary Data**: Puede mostrar datos binarios, como imágenes o archivos, en formato codificado.

* **Regular Expression**: Representa expresiones regulares.

* **UUID**: Representa un identificador único universal (UUID).

* **Decimal128**: Representa números decimales de precisión arbitraria.

# Sintaxis básica de consultas (`find()` y `distinct()`)

Estos métodos se utilizan para buscar documentos en una colección. MongoDB ofrece varios métodos para recuperar documentos de una colección que coincidan con un criterio especificado. Algunos de los métodos más comunes son:

- `find():` Este es el método básico utilizado para recuperar documentos de una colección. Toma un objeto de consulta como argumento y devuelve todos los documentos que coinciden con ese criterio.

- `find_one():` Similar a `find()`, pero devuelve solo el primer documento que coincida con el criterio especificado.

- `distinct():` Este método se utiliza para recuperar los valores únicos de un campo específico dentro de una colección. Retorna una lista de valores distintos para el campo especificado en los documentos de la colección.


## `find()`

Se utiliza para buscar documentos dentro de una colección que cumplan con ciertos criterios de consulta. Este método devuelve un cursor que contiene los documentos que coinciden con la consulta especificada.

La sintaxis básica del método es:

```pyhton
db.collection.find(query, projection)
```

- `db.collection`: Especifica la colección en la que se realizará la búsqueda.
  
- `query`: Especifica los criterios de búsqueda que deben cumplir los documentos para ser devueltos. Es opcional y si no se proporciona, devuelve todos los documentos de la colección.
  
- `projection`: Especifica los campos que se devolverán en los documentos coincidentes. También es opcional y si no se proporciona, se devolverán todos los campos.


> En nuestro caso, `db.collection` es igual a la variable que creamos previamente llamada `bcn`. 


````python
# empecemos con un ejemplo sencillo, sin introducir una query en el método 'find' y veámos que es lo que nos devuelve
bcn.find()
````

````python
# para convertir el cursor en información valiosa para nosotros podemos convertirlo a lista, como vemos a continuación
resultado_find_list = list(bcn.find())
print(resultado_find_list[:2])

# o lo podemos convertir a DataFrame directamente
resultado_find_df = pd.DataFrame(bcn.find())
display(resultado_find_df.head(2))
````

````python
# con esta query podremos hacer una análisis exploratorio inicial para conocer los principales estadísticos de nuestra BBDD de una forma sencilla
# Lo haremos primero para las variables numéricas
resultado_find_df.describe().T.iloc[4:, ]
````

````python
# hagamos lo mismo para las variables categóricas
resultado_find_df.describe(include = "O").T.iloc[1:, ]
````

## `find_one()`

Se utiliza para buscar un único documento dentro de una colección que cumpla con ciertos criterios de consulta. Este método devuelve el primer documento que coincide con la consulta especificada.

La sintaxis básica del método es similar a la del `find()`:

```python
db.collection.find_one(query, projection)
```

- `db.collection`: Especifica la colección en la que se realizará la búsqueda.
  
- `query`: Especifica los criterios de búsqueda que debe cumplir el documento para ser devuelto. Es opcional y si no se proporciona, devuelve el primer documento de la colección.
  
- `projection`: Especifica los campos que se devolverán en el documento coincidente. También es opcional y si no se proporciona, se devolverán todos los campos.

````python
# pongamos el mismo ejemplo que antes pero en este caso con el método ' find_one()' para ver las diferencias 
findone_stmarti_df =  pd.DataFrame(bcn.find_one({'neighbourhood_group': 'Gràcia'}), index=[0])

# comprobamos los resultados
print(f"Los distritos que tenemos en nuestra búsqueda son: {findone_stmarti_df['neighbourhood_group'].unique()} ")

# comprobemos el total de resultados obtenidos
print(f"El total de alojamientos es {findone_stmarti_df.shape[0]}")
````

## `distinct()`

Se utiliza para recuperar los valores únicos de un campo específico dentro de una colección. Esto significa que devuelve una lista de valores distintos para el campo especificado en los documentos de la colección.

```python
db.collection.distinct(query, projection)
```

- `db.collection`: Especifica la colección en la que se realizará la búsqueda.
  
- `query`: Especifica los criterios de búsqueda que debe cumplir el documento para ser devuelto. Es opcional y si no se proporciona, devuelve el primer documento de la colección.
  
- `projection`: Especifica los campos que se devolverán en el documento coincidente. También es opcional y si no se proporciona, se devolverán todos los campos.

> Importante, este método devuelve directamente una lista con los valores únicos del campo al que estamos accediendo. 

````python
# en este caso vamos a querer saber cuáles son los distintos distritos que tenemos dentro de nuestra BBDD, para eso usaremos la siguiente query
bcn.distinct('neighbourhood_group')
````

````python
# también podríamos saber el número de barrios distintos que hay en Barcelona
len(bcn.distinct("neighbourhood"))
````

````python
# imaginamos ahora que queremos saber cuántas números de reviews distintas hay en la BBDD
bcn.distinct("number_of_reviews")[:5]
````

# Operadores de Consulta

Los operadores de consulta permiten realizar búsquedas avanzadas dentro de una colección de documentos. Estos operadores se utilizan en combinación con los métodos `find()` y `find_one()` para filtrar documentos en función de ciertos criterios.
| **Categoría**               | **Operador**   | **Descripción**                                                                                                      | **Ejemplo**                                                                                                                                         |
|-----------------------------|----------------|----------------------------------------------------------------------------------------------------------------------|-----------------------------------------------------------------------------------------------------------------------------------------------------|
| **Operadores de comparación**| `$eq`          | Igual a un valor específico.                                                                                         | `{ campo: { $eq: valor } }`                                                                                                                         |
|                             | `$ne`          | No igual a un valor específico.                                                                                      | `{ campo: { $ne: valor } }`                                                                                                                         |
|                             | `$gt`          | Mayor que un valor específico.                                                                                       | `{ campo: { $gt: valor } }`                                                                                                                         |
|                             | `$gte`         | Mayor o igual a un valor específico.                                                                                 | `{ campo: { $gte: valor } }`                                                                                                                        |
|                             | `$lt`          | Menor que un valor específico.                                                                                       | `{ campo: { $lt: valor } }`                                                                                                                         |
|                             | `$lte`         | Menor o igual a un valor específico.                                                                                 | `{ campo: { $lte: valor } }`                                                                                                                        |
| **Operadores lógicos**       | `$and`         | Realiza una operación lógica "AND" en una lista de expresiones.                                                      | `{ $and: [ { expresion1 }, { expresion2 }, ... ] }`                                                                                                 |
|                             | `$or`          | Realiza una operación lógica "OR" en una lista de expresiones.                                                       | `{ $or: [ { expresion1 }, { expresion2 }, ... ] }`                                                                                                  |
|                             | `$nor`         | Realiza una operación lógica "NOR" en una lista de expresiones.                                                      | `{ $nor: [ { expresion1 }, { expresion2 }, ... ] }`                                                                                                 |
|                             | `$not`         | Niega una expresión de consulta.                                                                                     | `{ campo: { $not: { $eq: valor } } }`                                                                                                               |
| **Operadores de elemento**   | `$exists`      | Coincide con los documentos que contienen un campo específico.                                                       | `{ campo: { $exists: true } }`                                                                                                                      |
|                             | `$type`        | Coincide con los documentos que tienen un tipo de campo específico.                                                  | `{ campo: { $type: tipo } }`                                                                                                                        |
| **Otros operadores**         | `$in`          | Coincide con los documentos cuyo valor de campo está en una lista de valores.                                        | `{ campo: { $in: [valor1, valor2, valor3] } }`                                                                                                      |
|                             | `$nin`         | Coincide con los documentos cuyo valor de campo no está en una lista de valores.                                     | `{ campo: { $nin: [valor1, valor2, valor3] } }`                                                                                                     |
|                             | `$regex`       | Realiza una búsqueda de expresiones regulares en los campos de texto.                                                | `{ campo: { $regex: /expresion_regular/ } }`                                                                                                        |


Pongamos algunos ejemplos de cada uno de estos operadores que nos van a permitir filtrar la información de la BBDD para columnas de tipo numérico. 

````python
# en el análisis preliminar habíamos visto que el precio más barato de un alojamiento en Barcelona era de 9€, vamos a identificar que alojamiento/s son usando el operador '$eq'
find_eq_df = pd.DataFrame(bcn.find({'price': {"$eq": 9}}))
find_eq_df.head()
````

````python
# imaginemos ahora que queremos encontrar todos los alojamientos que no tienen el precio mínimo identificado en pasos previos, 
# tendríamos que usar el operador '$ne' 
find_ne_df = pd.DataFrame(bcn.find({'price': {"$ne": 9}}))
find_ne_df.head()
````

````python
# imagemos que queremos conocer los alojamientos que tienen un precio mayor a 100, para eso tendremos que usar el operadore '$eq'
find_gt_df = pd.DataFrame(bcn.find({'price': {"$gt": 1000}}))
display(find_gt_df.head())

print(f"Tenemos un total de {find_gt_df.shape[0]}, que están en los siguientes distritos {find_gt_df['neighbourhood_group'].unique()}")
````

````python
# antes hemos identificado los barrios y alojamientos que tienen un precio mayor de 1000 €, si quisieramos extraer los que tengan un valor mayor o igual usaremos el operador '$gte'
# en este caso vamos que 14 barrios nuevos, los cuales nos podrían interesar identificar
find_gte_df = pd.DataFrame(bcn.find({'price': {"$gte": 1000}}))
display(find_gte_df.head())

print(f"Tenemos un total de {find_gte_df.shape[0]}, que están en los siguientes distritos {find_gte_df['neighbourhood_group'].unique()}")

print(f" \n Los distritos que tienen alquileres a 1000€ son {set(find_gte_df['neighbourhood_group'].unique()) - set(find_gt_df['neighbourhood_group'].unique())}")
````

````python
# hagamos este mismo ejercicio poniendo el limite superior en 35€ (es el precio medio que se gasta un español en alojamiento cuando sale de vacaciones)
# tendremos que usar el operador '$lt'

find_lt_df = pd.DataFrame(bcn.find({"price": {"$lt": 35}}))

display(find_lt_df.head())

print(f"Tenemos un total de {find_lt_df.shape[0]}, que están en los siguientes distritos {find_lt_df['neighbourhood_group'].unique()}")
````

````python
# hagamos lo mismo pero con el operador '$lte'
find_lte_df = pd.DataFrame(bcn.find({'price': {"$lte": 35}}))
display(find_lte_df.head())

print(f"Tenemos un total de {find_lte_df.shape[0]}, que están en los siguientes distritos {find_lte_df['neighbourhood_group'].unique()}")

print(f" \n Los distritos que tienen alquileres a 35 son {set(find_lte_df['neighbourhood_group'].unique()) - set(find_lt_df['neighbourhood_group'].unique().tolist())}")
````

````python
# lo siguiente pregunta que se nos puede ocurrir es si hay algun distrito que esté entre los distritos más baratos y más caros
# para eso podríamos hacer lo siguiente
set(find_gte_df["neighbourhood_group"].unique()).intersection(set(find_lte_df["neighbourhood_group"].unique()))
````

## Operadores lógicos

Son utilizados para realizar operaciones de comparación entre expresiones y documentos en las consultas. Los operadores lógicos más comunes son: 

| **Operador** | **Descripción**                                                                                                                      | **Ejemplo**                                                                                                            |
|--------------|--------------------------------------------------------------------------------------------------------------------------------------|------------------------------------------------------------------------------------------------------------------------|
| **$and**     | Selecciona documentos que coincidan con todas las condiciones especificadas. Todas las condiciones deben ser verdaderas.              | `db.collection.find({ '$and': [ { <condicion1> }, { <condicion2> } , ... , { <condicionN> } ]})`                       |
| **$or**      | Selecciona documentos que coincidan con al menos una de las condiciones especificadas. Al menos una condición debe ser verdadera.     | `db.collection.find({ '$or': [ { <condicion1> }, { <condicion2> } , ... , { <condicionN> } ]})`                        |
| **$not**     | Invierte el efecto de una expresión de consulta y selecciona documentos que no coincidan con las condiciones especificadas.           | `db.collection.find({ '$not': [ { <condicion1> }, { <condicion2> } , ... , { <condicionN> } ]})`                       |
| **$nor**     | Selecciona documentos que no coincidan con ninguna de las condiciones especificadas. Ninguna de las condiciones debe ser verdadera.    | `db.collection.find({ '$nor': [ { <condicion1> }, { <condicion2> } , ... , { <condicionN> } ]})`                       |

````python
# imaginemos que queremos saber si hay algún alojamiento que tiene un precio igual o menos a 35€ y con un total de camas de 6
# para eso tendríamos que usar el operdor '$and' porque queremos que se cumplan las dos condiciones 

# empezamos definiendo las condiciones  que queremos evaluar
cond_precio = {"price": {"$lte": 35}}
cond_camas = {"rooms_number": "6 bedrooms"}

# realizamos la query
find_and_df = pd.DataFrame(bcn.find({"$and": [cond_precio, cond_camas]}))
find_and_df
````

````python
# definimos las dos condiciones
cond_tipo = {"room_type": "Shared room"}
cond_reviews = {"number_of_reviews": {"$eq": 0}}

# realizamos la query
find_or_df = pd.DataFrame(bcn.find({"$or": [cond_tipo, cond_reviews]}))
find_or_df.head(2)
````

````python
# si quisieramos hacer la inversa de '$or', podríamos usar el método '$nor', 
find_nor_df = pd.DataFrame(bcn.find({"$nor": [cond_tipo, cond_reviews]})) 
display(find_nor_df.head(2))

# en este caso si sumamos el número de filas del resultado de la query del 'or' y de la query del 'nor' nos devolvería el total del filas del conjunto original
find_or_df.shape[0] + find_nor_df.shape[0] == resultado_find_df.shape[0]
````

## Operadores de elemento

Son utilizados para realizar consultas basadas en la existencia de un campo o atributo en un documento, así como para verificar si un campo tiene un valor específico o si es de un tipo de dato particular. Los métodos más comunes son: 

1. **$exists:**: Se utiliza para verificar si un campo existe en un documento.
   
   ```python
   db.collection.find({campo: {'$exists': True|False}})
   ```
   Donde:

   - **db.collection:** Es el nombre de la colección de nuestra BBDD.

   - **campo:** Nombre del campo que queremos indicar si existe registro o no. Especificaremos *True* si queremos obtener los documentos que tengan ese campo relleno y, por el contrario, usaremos *False* si queremos que en ese campo haya un nulo.


2. **$type:**: Se utiliza para verificar si un campo tiene un tipo de dato específico.
   
   ```python
   list(db.collection.find({campo:{'$type': 'tipo'}}))
   ```
   Donde:

   - **db.collection:** Es el nombre de la colección de nuestra BBDD.
  
   - **campo:** Nombre del campo que queremos verificar la instancia de tipo concreto.
  
   - **tipo:** Tipode dato que queremos buscar. 

   Este tipo de operador lo podremos usar para la validación de datos para verificar si los campos de interés tienen el tipo de datos esperado, para identidicar campos con tipos de datos incorrectos o inesperados o para seleccionar campos con un tipo de dato específico, como por ejemplo los nulos.

````python
# imaginemos ahora que queremos saber cuáles son aquellos alojamientos que no tienen ninguna estrella ya que queremos evitarlos a la hora de hacer recomendaciones.
# para eso podemos usar el operador '$exists'  para seleccionar aquellos alojamientos que no tienen estrellas 
# fijaos como ahora no tenemos la columna de 'stars' en el DataFrame resultante
find_exists_df = pd.DataFrame(bcn.find({'stars': {'$exists': False}}))
print(f"Tenemos un total de {round((find_exists_df.shape[0] / resultado_find_df.shape[0]), 2) * 100} % alojamientos sin estrellas")
````

````python
# siguiendo en la línea del ejemplo anterior, imaginemos que no queremos tener en cuenta aquellos alojamientos que no tienen precio o que no ha recibido ninguna puntuación para evitar recomendarlos
# NOTA: este mismo ejemplo lo podríamos hacer también con el '$exists'

# definimos las condiciones
cond_estrellas = {"stars": {"$exists": False}}
cond_precio = {"price": {"$exists": False}}

# aplicamos el operador '$and' para combinar las dos condiciones
find_type_df = pd.DataFrame(bcn.find({'$and': [cond_estrellas, cond_precio]}))

print(f"Tenemos un total de {round(find_type_df.shape[0] / resultado_find_df.shape[0], 2) * 100} % alojamientos sin estrellas ni precio")
````

## Otros operadores

En Mongo tenemos otros operadores a parte de los mencionados hasta este punto, los más importantes son: 

1. **$in:** Se utiliza para verificar si el valor de un campo está dentro de una lista de valores especificada.
   
   ```python
   list(db.collection.find({'nombre del campo': {'$in': [valor1, valor2, ...]}}))
   ```

   - **nombre del campo:** Campo del que queremos obtener los valores.

   ```python

   ```
   Donde:

   - **db.collection:** Es el nombre de la base de datos y la colección con la que estamos trabajando.
  
   - **nombre del campo:** Campo del que queremos obtener los valores.
  
   - **operador**: Operador numérico que vamos a querer utilizar.

   - **[valor1, valor2, ...]**: Lista de valores  que vamos a usar de referencia para filtrar la información. 
  

3. **$all:** Se utiliza para verificar si un campo contiene **todos** los valores especificados en una lista.
   
   ```python
   list(db.collection.find({'nombre del campo': {'$all': [valor1, valor2, ...]}}))
   ```
   Donde:

   - **db.collection:** Es el nombre de la base de datos y la colección con la que estamos trabajando.
  
   - **nombre del campo:** Campo del que queremos obtener los valores.
  
   - **operador**: Operador numérico que vamos a querer utilizar.

   - **[valor1, valor2, ...]**: Lista de valores  que vamos a usar de referencia para filtrar la información. 
  

4. **$size:** Se utiliza para verificar si un campo es una lista y tiene un tamaño específico.

   ```python
   list(db.collection.find({'nombre del campo': {'$size': numero}}))
   ```

   Donde:

   - **db.collection:** Es el nombre de la base de datos y la colección con la que estamos trabajando.
  
   - **nombre del campo:** Campo del que queremos obtener los valores.
  
   - **operador**: Operador numérico que vamos a querer utilizar.

   - **numero**: Número de elementos que debe tener la lista del campo. 
  

5. **$regex:** Se utiliza para realizar una búsqueda de expresiones regulares en el contenido de un campo.

   Donde:

En Mongo tenemos otros métodos que nos puedes ayudar a ordernar o limitar los resultados de nuestras queries. Los más importantes son:

2. **limit()**: Limita la cantidad de documentos devueltos por una consulta. Es como el `limit` de SQL.

## `sort()`

Se utiliza para ordenar los resultados de una consulta en función de los valores de un campo específico. Esto es útil cuando queremos obtener los documentos de una colección en un orden específico, como ordenarlos alfabéticamente, por fecha, o por algún otro criterio relevante para tu aplicación. Podemos ordenar por múltiples campos especificándolos dentro del objeto de ordenamiento. En caso de empate en el primer campo, se utilizará el segundo campo para desempatar y así sucesivamente.


La sintaxis básica es:

```python
db.collection.find({ <filtro> }).sort({ <campo>: <1 o -1> })
```
Donde: 
- `db.collection`: Es la colección en la que deseas realizar la consulta.

- `find({ <filtro> })`: Es la consulta que especifica los documentos que deseas recuperar de la colección.

- `.sort({ <campo>: <1 o -1> })`: Es el método `sort()` que ordena los documentos basados en el valor del campo especificado. Puedes proporcionar uno o varios campos como criterio de ordenamiento.

   - `<campo>`: Es el nombre del campo por el cual deseas ordenar los documentos.

   - `<1 o -1>`: Especifica el orden de clasificación. Utiliza `1` para orden ascendente (de menor a mayor) y `-1` para orden descendente (de mayor a menor).

Se utiliza  para limitar el número de documentos que se devuelven en el resultado de una consulta. Podemos usar `limit()` para restringir la cantidad de documentos devueltos por una consulta a una colección. El método `limit()` se aplica después de que se hayan seleccionado y ordenado los documentos, por lo que limita la cantidad de documentos devueltos después de que se hayan aplicado todas las otras operaciones de la consulta. Si no se especifica ningún valor para `limit()`, la consulta devolverá todos los documentos que cumplan con los criterios de la consulta sin ninguna restricción en la cantidad de documentos devueltos. 

La sintaxis básica del método es:

```python
db.coleccion.find(query).limit(numero)
```

- `db.coleccion`: Especifica la colección en la que se realizará la consulta.

- `query`: Especifica las condiciones de la consulta para filtrar los documentos.

- `.limit(numero)`: Especifica el número máximo de documentos que se devolverán en el resultado de la consulta.

# Agregaciones en Mongo

Las agregaciones son operaciones que permiten procesar datos de forma dinámica y flexible en una colección, utilizando una variedad de etapas que se aplican secuencialmente a los documentos. Estas etapas pueden realizar diversas operaciones, como filtrado, agrupación, proyección, ordenamiento, y aplicar funciones de agregación (por ejemplo, sumar, contar, promediar, etc.).

Las agregaciones son extremadamente útiles para los analistas de datos por varias razones:

- **Análisis avanzado de datos**: Las agregaciones permiten realizar análisis complejos y avanzados en los datos almacenados en una base de datos. Podremos realizar operaciones como agrupar datos, calcular estadísticas, y realizar cálculos basados en múltiples campos.

- **Flexibilidad**: Las agregaciones son altamente flexibles y pueden adaptarse a una amplia gama de requisitos de análisis. 

- **Rendimiento**: Las agregaciones pueden ejecutarse de manera eficiente incluso en grandes conjuntos de datos, lo que permite realizar análisis rápidos y eficientes en tiempo real.

Las principales agregaciones que tenemos en Mongo son: 


1. **$match**: Filtra los documentos de la colección según un criterio específico. Es similar al `find()` que hemos estado viendo hasta ahora.

   ```python
   { $match: { campo: valor } }
   ```

2. **$group**: Agrupa los documentos según un campo o una expresión y calcula estadísticas sobre los documentos agrupados. Es como el groupby de SQL que conocemos. 

   ```python
   { $group: { _id: "$campo", nueva_col: { $sum: 1 } } }
   ```

3. **$project**: Proyecta campos específicos de los documentos y puede incluir/excluir campos, crear nuevos campos calculados o renombrar campos existentes. Igual que estuvimos viendo hasta ahora. 

   ```python
   { $project: { campo: 1, _id: 0 } }
   ```

4. **$sort**: Ordena los documentos según un campo específico, en orden ascendente o descendente.

   ```python
   { $sort: { campo: 1 } }
   ```

5. **$limit**: Limita el número de documentos que se pasan a la siguiente etapa de la agregación.

   ```python
   { $limit: 10 }
   ```

6. **$skip**: Omite un número específico de documentos antes de pasar al siguiente etapa de la agregación.

   ```python
   { $skip: 5 }
   ```

7. **$lookup**: Realiza una "unión" entre dos colecciones, recuperando documentos de una colección que coinciden con los documentos de la colección actual. Podríamos compararlo con los joins de SQL. 

   ```python
   {
     $lookup:
       {
         from: "otra_coleccion",
         localField: "campo_local",
         foreignField: "campo_foraneo",
         as: "nombre_nuevo"
       }
   }
   ```




````python
# imaginemos ahora que queremos saber los 5 anfitriones con más alojamientos en la base de datos
# como hemos hablado, esto es lo que en SQL sería un groupby. 

query_top_5 = [
    #  Aquí estamos agrupando por la columna 'host_id' (esto lo especificamos usando la clave '_id')
    # en este caso, en lugar de sumar un valor numérico específico, se utiliza `$sum: 1`. Esto significa que, en cada grupo generado por `host_id`, se contará el número de documentos en ese grupo.
    { "$group": { "_id": "$host_id", "total_alojamientos": { "$sum": 1 } } },  

    # aquí estamos eligiendo las columnas que querremos en nuestros resultados
    { "$project": { "_id": 1, "total_alojamientos": 1 , "host_name": 1} },  

    # por último ordenamos los resultados de mayor a menor en base a la columna creada en el agrupamiento 
    { "$sort": { "total_alojamientos": -1 } }, 

    # limitamos los resultados a 5
    {"$limit": 5}
]

# ejecutamos la consulta
df_top_anfi = pd.DataFrame(bcn.aggregate(query_top_5))

# mostramos los resultados
df_top_anfi
````

````python
# la query anterior estar bien, pero los id_ de los anfitriones no nos aportan mucho valor, en esta query vamos a añadir el nombre de los anfitriones 
query_top_5_2 = [
    # Agrupar por host_id y obtener el nombre del anfitrión
    {
    "$group": { # iniciamos el proceso de agrupación
            "_id": "$host_id", #  Especificamos el campo por el cual queremos agrupar los documentos.  
            "anfitrion": { "$first": "$host_name" }, # El operador $first se utiliza para seleccionar el primer valor de host_name encontrado en cada grupo.
            "total_alojamientos": { "$sum": 1 } # Calcular el número total de alojamientos en cada grupo. 
             }
    },  

    # Proyectar los campos deseados en los resultados
    { "$project": { "_id": 1, "total_alojamientos": 1, "anfitrion": 1 } },  

    # Ordenar los resultados de mayor a menor en base al número total de alojamientos
    { "$sort": { "total_alojamientos": -1 } }, 

    # Limitar los resultados a 5
    {"$limit": 5}
]

df_top_afi2 = pd.DataFrame(bcn.aggregate(query_top_5_2))
df_top_afi2
````

````python
# en este caso queremos saber cual es el precio medio de los alojamientos por distrito
query_precio_medio = [
    #Agrupamos los datos por distrito, creando una columna nueva llamada "precio_promedio" resultado de sacar el precio medio por distrito
    { "$group": { "_id": "$neighbourhood_group", "precio_promedio": { "$avg": "$price" } } } , 

    # ordenamos los resultados por el campo nuevo 'precio_promedio' de mayor a menor
    {"$sort": {"precio_promedio": -1}}
]
      

# Ejecutar la consulta de agregación
df_precio_medio = pd.DataFrame(bcn.aggregate(query_precio_medio)).round(2)
df_precio_medio
````

````python
# compliquemos un poco más la query anterior y supongamos que no queremos sacar la media de todos los distritos, que solo nos interesan los distitos de 'Les Corts', 'Eixample' y 'Gràcia'
# esto lo podríamos hacer filtrando el DataFrame final, pero Mongo nos permite utilizar la sentencia "$match" para filtrar los datos que luego queremos agrupar


# Consulta de agregación para calcular la media de precios por barrio
query_media_precio2 = [

    # filtramos los datos por unos distritos concretos que nos interesan
    { "$match": 
            { "neighbourhood_group": {"$in": ['Les Corts', 'Eixample' , 'Gràcia']} } 
    }, 

    # agrupamos los datos en base al campo de distrito y creamos una columna nueva llamada 'precio_promedio' resultado de calcular la media del precio
    { "$group": 
            { "_id": '$neighbourhood_group', "precio_promedio": { "$avg": "$price" } } 
    } 
]

# Ejecutar la consulta de agregación
df_precio_medio2 = pd.DataFrame(bcn.aggregate(query_media_precio2))
df_precio_medio2
````

En esta query de MongoDB estamos realizando una operación de agregación utilizando el operador `$lookup`, seguido de una etapa de proyección con el operador `$project`.


   - `from`: Especificamos la colección con la que queremos realizar la unión, en este caso, la colección "Madrid".

   - `localField`: Especificamos el campo de la colección actual (`host_id`) que se utilizará para hacer la unión.

   - `foreignField`: Especificamos el campo de la colección especificada en `from` (`host_id`) que se utilizará para hacer la unión.

   - `as`: Especificamos el nombre del campo que contendrá los resultados de la unión en los documentos de la colección actual. En este caso, se utiliza el alias "alojamientos_madrid".

   - `$project`: Después de la unión, aplicamos una etapa de proyección para incluir solo los campos específicos en la salida. En esta etapa:


   - `{"alojamientos_madrid": 1}`: Incluimos solo el campo "alojamientos_madrid" en la salida, con un valor de `1` que indica que se debe incluir en la salida.

En resumen, esta query realiza una unión entre la colección actual y la colección "Madrid" en base al campo `host_id`, y luego proyecta solo el campo "alojamientos_madrid" en la salida.

````python
pipeline = [
    { # realiza una "unión" entre dos colecciones
        "$lookup": {
            # Especificamos la colección con la que queremos realizar la unión
            "from": "Madrid",

            # Campo de unión en la colección de 'Madrid'
            "localField": "host_id",  

            # Campo de unión en la colección 'Barcelona'
            "foreignField": "host_id",  

            # Alias para los documentos de la colección 'Madrid'
            "as": "alojamientos_madrid"  
        }
    }, 
    {
        "$project": {
            "alojamientos_madrid": {
                "$mergeObjects": "$alojamientos_madrid"  # Incluye todos los campos de los documentos originales
            }
        }
    }
]

resultados = pd.DataFrame(bcn.aggregate(pipeline))

# mostramos 3 resultados aleatorios
resultados.sample(3)
````

Estas fechas se almacenan en MongoDB con el tipo de dato *datetime*, por ello es imprescindible utilizar la librería *datetime* de python.

MongoDB ofrece una amplia gama de operadores y funciones para realizar consultas basadas en fechas. # Índices

Los índices son estructuras de datos que mejoran la eficiencia de las consultas al permitir un acceso más rápido a los datos. MongoDB ofrece una variedad de tipos de índices y estrategias de indexación para adaptarse a diferentes situaciones y consultas. La importancia de estos índices radica en: 

- Como ya hemos dicho, mejoran el rendimiento de las consultas al acelerar la recuperación de datos.

- Ayudan a optimizar las operaciones de lectura al reducir la cantidad de documentos que deben ser escaneados.

- Facilitan la búsqueda y clasificación de datos, especialmente en grandes colecciones.

Los tipos de índices que tenemos en Mongo son: 


- **Índices Simples**: Estos son los índices más básicos en MongoDB y se crean en un solo campo. Son útiles para mejorar el rendimiento de las consultas que filtran datos basados en un solo campo.

    ```python
    # Ejemplo de creación de un índice simple en un campo "edad"
    db.collection.create_index("edad")
    ```

- **Índices Compuestos**: Estos índices se crean en múltiples campos y son útiles para consultas que filtran o ordenan datos basados en múltiples campos.

    ```python
    # Ejemplo de creación de un índice compuesto en dos campos "nombre" y "apellido"
    db.collection.create_index([("nombre", pymongo.ASCENDING), ("apellido", pymongo.ASCENDING)])
    ```

- **Índices de Texto**: Estos índices se utilizan para realizar búsquedas de texto completo en campos de tipo cadena de caracteres. Son útiles para realizar consultas de búsqueda de texto en grandes conjuntos de datos.

    ```python
    # Ejemplo de creación de un índice de texto en un campo "descripcion"
    db.collection.create_index([("descripcion", "text")])
    ```

- **Índices Geoespaciales**: Estos índices se utilizan para consultas espaciales en datos que contienen información geoespacial, como coordenadas de ubicación. Son útiles para analizar datos basados en su ubicación geográfica.

    ```python
    # Ejemplo de creación de un índice geoespacial en un campo "ubicacion"
    db.collection.create_index([("ubicacion", pymongo.GEOSPHERE)])
    ```

- **Índices de Hash**: Estos índices se utilizan para mejorar el rendimiento de las consultas que realizan búsquedas en campos que contienen datos hash, como hashes de contraseñas o tokens de autenticación.

    ```python
    # Ejemplo de creación de un índice de hash en un campo "contraseña"
    db.collection.create_index([("contraseña", pymongo.HASHED)])
    ```

# Consultas Geoespaciales

En MongoDB, las consultas geoespaciales permiten realizar operaciones espaciales en datos que contienen información de ubicación geográfica. Esto es especialmente útil cuando se trabaja con conjuntos de datos que incluyen coordenadas geográficas, como puntos de interés, ubicaciones de clientes o datos de seguimiento de vehículos.  Nos va a permitir encontrar ubicaciones cercanas a punto específico, calcular distancias entre puntos, y realizar análisis espaciales avanzados. 

Los principales métodos de consulta geoespacial que tenemos en Mongo son: 


1. **Near**: Este método se utiliza para encontrar documentos cerca de un punto geoespacial específico. Se puede configurar para buscar documentos dentro de una distancia máxima definida del punto de referencia.

2. **GeoNear**: Este método nos permite realizar consultas que devuelven documentos ordenados por proximidad a un punto de referencia. A diferencia con el método `near` nos va a devolver la distancia que hay desde cada uno de los documentos al punto de referencia. 

2. **geoWithin**: Permite buscar documentos que se encuentren dentro de una forma geoespacial específica, como un polígono o un círculo. Esto es útil para encontrar ubicaciones dentro de áreas específicas definidas por coordenadas geográficas.

3. **geoIntersects**: Este método se utiliza para buscar documentos cuyas formas geoespaciales intersecten con una forma geoespacial específica. Puede ser útil para encontrar documentos que se superponen con áreas geográficas específicas.

4. **GeoWithin**: Similar al método `Within`, pero optimizado para buscar documentos dentro de un polígono definido por coordenadas geográficas.

5. **GeoIntersects**: Similar al método `Intersects`, pero optimizado para buscar documentos que intersecten con una forma geoespacial definida.


Al trabajar con consultas geoespaciales en MongoDB, es importante tener en cuenta la estructura de los datos geoespaciales, la indexación adecuada para mejorar el rendimiento de las consultas y las proyecciones para seleccionar y mostrar información relevante. Además, es fundamental comprender los distintos métodos de consulta geoespacial y cómo se pueden aplicar en diferentes escenarios de aplicación.


## `near()`

Se utiliza para buscar documentos que están cerca de un punto geográfico específico en una colección que contiene datos geoespaciales. Este método es útil para encontrar lugares cercanos a una ubicación dada, como restaurantes cercanos, tiendas o puntos de interés.

La sintaxis básica es:

```python
db.collection.find({
    location_field: {
        $near: {
            $geometry: {
                type: "Point",
                coordinates: [longitude, latitude]
            },
            $maxDistance: max_distance_in_meters
        }
    }
})
```
Donde: 

- `collection`: El nombre de la colección en la que se realizará la búsqueda.

- `location_field`: El nombre del campo que contiene los datos geoespaciales.

- `$near`: El operador que indica que se realizará una búsqueda de documentos cerca de un punto geográfico.

- `$geometry`: Un objeto que define el tipo de geometría y las coordenadas del punto de referencia.

- `type`: El tipo de geometría, que generalmente es `"Point"` para coordenadas geográficas.

- `coordinates`: Un array que especifica las coordenadas `[longitud, latitud]` del punto de referencia.

- `$maxDistance`: La distancia máxima en metros dentro de la cual se buscarán los documentos.

<div style="background-color: #F74646; padding: 10px; color: black;">
Cuando trabajamos con geoqueries, tendremos que hacer dos cosas: 

- En primer lugar, es necesario convertir los campos de latitud y longitud a un formato de geometría específico, como el formato GeoJSON, para realizar consultas. Esto es así ya que Mongo utiliza un sistema de índices geoespaciales para optimizar las consultas basadas en la ubicación y para que pueda interpretar y procesar estas consultas correctamente. El formato que debe de tener el nuevo campo que creemos es: 

    ```python
    "geometry": {"type": "Point",
                "coordinates": ["$latitude", "$longitude"]}
    ```

- En segundo lugar, es esencial crear un índice `2dsphere` en los campos de ubicación. Sin este índice, MongoDB tendría que explorar toda la colección para cada consulta geoespacial, lo que sería muy ineficiente, especialmente en colecciones grandes. 
</div>


````python
# Vamos con el primer paso, la crearción del campo "geometry" 
# Actualizamos todos los documentos de la colección utilizando el método 'update_many()' 
# Añadimos un nuevo campo llamado 'geometry' a todos los documentos de tipo 'Point' con las coordenadas que tenemos en la BBDD

actualizacion = bcn.update_many(
    {},  # Filtro vacío para que coincida con todos los documentos en la colección
    [
        {
            "$set": { # usamos el método set para crear un campo nuevo en todos los documentos
                "geometry": {  # Especificamos el nombre del nuevo campo que queremos crear
                    "type": "Point",  # Tipo de geometría
                    "coordinates": ["$latitude", "$longitude"]  # Coordenadas de longitud y latitud, utilizando las columnas que tenemos en neustra base de datos
                }
            }
        }
    ]
)

# Comprobamos que todos los campos se han actualizado con  éxito
print("Documentos actualizados:", actualizacion.modified_count)
````

````python
# Vamos ahora con el segundo paso, la creación del índice 2dsphere para poder hacer las geoqueries
# para eso usaremos el método 'create_index()'  que aprendimos en el apartado anterior.
bcn.create_index([("geometry", '2dsphere')])
````

## `geoNear()`

El método `geoNear` se utiliza para realizar consultas geoespaciales que devuelven documentos ordenados por proximidad a un punto de referencia. Su sintaxis básica es la siguiente:

```python
db.collection.geoNear({
    near: { type: "Point", coordinates: [longitud, latitud] },
    spherical: True,
    maxDistance: distancia_maxima,
    query: { <opcional: filtros adicionales> },
    distanceField: "<opcional: campo para almacenar la distancia>", 
    distanceMulyiplier: 1
})
```

- `near`: Especifica el punto de referencia para la consulta, con las coordenadas de longitud y latitud.

- `spherical`: Indica si se debe usar la geometría esférica para calcular distancias en la Tierra.

- `maxDistance`: Opcionalmente, establece la distancia máxima en metros desde el punto de referencia.

- `query` (opcional): Permite aplicar filtros adicionales a los documentos antes de calcular la proximidad.

- `distanceField` (opcional): Define el nombre del campo donde se almacenará la distancia del documento al punto de referencia.

- `distanceMultiplier` (opcional): Se utiliza para ajustar la distancia calculada entre los documentos y el punto de referencia especificado. Por defecto, MongoDB devuelve la distancia en metros entre los documentos y el punto de referencia. Sin embargo, si se especifica un valor para `distanceMultiplier`, MongoDB multiplicará la distancia calculada por este valor antes de devolverla. Esto puede ser útil para convertir la distancia a una unidad diferente, como kilómetros o millas, según sea necesario para la aplicación. Por ejemplo, si se establece `distanceMultiplier` en `0.001`, la distancia se devolverá en kilómetros en lugar de metros.

## geoWithin

Se utiliza para buscar documentos dentro de una forma geoespacial específica, como un polígono o un círculo. Permite filtrar los documentos de una colección según su ubicación en relación con una región geoespacial predefinida.

```python
db.collection.find({
```
