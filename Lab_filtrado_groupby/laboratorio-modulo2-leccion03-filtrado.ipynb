```python
import pandas as pd
import numpy as np
```

```python
# Carga el dataset de trabajo
netflix_filtrado = pd.read_csv('datos/laboratorio-modulo2-leccion02-eda-unionfinal.csv', index_col=0)
netflix_filtrado.columns
```

#### Ejercicio 1: ¿Cuáles son todos los títulos lanzados en 2021 que son del tipo 'Movie'?

```python
netflix_filtrado[(netflix_filtrado['type'] == 'Movie') & (netflix_filtrado['release_year'] == 2021)].head(5)
```

#### Ejercicio 2: ¿Qué títulos tienen una duración de 90 minutos? Muestra solo las columnas de título y de duración.

```python
condicion_duracion = netflix_filtrado['Runtime'] == 90
netflix_filtrado.loc[condicion_duracion,['Title','Runtime']]
```

#### Ejercicio 3: ¿Cuáles son todos los títulos del país 'India' lanzados después de 2018?

```python
netflix_filtrado[(netflix_filtrado['country'] == 'India') & (netflix_filtrado['release_year'] > 2018)].head(5)
```

#### Ejercicio 4: ¿Qué títulos tienen clasificación 'TV-14' o 'R'?

```python
netflix_filtrado[(netflix_filtrado['rating'] == 'TV-14') | (netflix_filtrado['rating'] == 'R')].head(5)
```

#### Ejercicio 5: ¿Cuáles son los títulos dirigidos por 'Christopher Nolan' o 'Martin Scorsese'? ¿Cuántos a dirigido cada uno de los directores?

```python
netflix_filtrado[(netflix_filtrado['director'] == 'Christopher Nolan') | (netflix_filtrado['director'] == 'Martin Scorsese')]
```

Solo hay una película de Martin Scorsese

#### Ejercicio 6: ¿Qué títulos pertenecen a los países 'United States', 'United Kingdom' o 'Canada'? Dentro de estos países, ¿cuántas películas tienen más de un lenguaje?

```python
paises = ['United States', 'United Kingdom', 'Canada']
paises_merged = netflix_filtrado[netflix_filtrado['country'].isin(paises)]
paises_merged[['country', 'Title']]
```

```python
filtro = paises_merged["Language"].str.contains("/")
peliculas_plurilen = paises_merged[filtro]
total_peliculas = peliculas_plurilen.shape[0]
total_peliculas
```

#### Ejercicio 7: ¿Qué títulos fueron lanzados entre 2015 y 2020, inclusive?

```python
netflix_filtrado.dtypes
```

```python
filtro_between = netflix_filtrado['release_year'].between(2015, 2020, inclusive='both')
netflix_filtrado[filtro_between]
```

#### Ejercicio 8: ¿Cuáles son los títulos cuya puntuación en IMDb está entre 6.5 y 8.0 y que no sepamos cual es su duración?

```python
filtro_imdb = netflix_filtrado['IMDB Score'].between(6.5, 8.0, inclusive='both')
filtro_duracion = netflix_filtrado['duration'].isna()
netflix_filtrado[filtro_imdb & filtro_duracion]
```

#### Ejercicio 9: ¿Cuáles son las películas cuya duración de 90 o 150 minutos?

```python
netflix_filtrado[(netflix_filtrado['Runtime'] == 90) | (netflix_filtrado['Runtime'] == 150)]
```

#### Ejercicio 10: ¿Cuáles son los títulos que contienen la palabra 'Love' en su título y que sean de tipo TV Show?

```python
patron_regex = 'love'
filtro_regex = netflix_filtrado['Title'].str.contains(patron_regex, case = False, na= False) ### decide si es sensible o no a las mayúsculas
netflix_love = netflix_filtrado[filtro_regex]
netflix_love
```

#### Ejercicio 11: ¿Cuáles son los títulos en los que el director tiene el apellido 'Johnson'?

```python
patron_director = 'Johnson'
filtro_director1 = netflix_filtrado['director'].str.contains(patron_director, case = False, na= False) ### decide si es sensible o no a las mayúsculas
netflix_filtrado[filtro_director1]
```

#### Ejercicio 12: ¿Tenemos algún documental ('Documentaries') de Mike Tyson?

#### Ejercicio 17: Encuentra los 10 títulos con las puntuaciones más altas en IMDb.

```python
netflix_top10 = netflix_filtrado.nlargest(10, 'IMDB Score').filter(['IMDB Score', 'Title'])
netflix_top10
```

#### Ejercicio 18: Encuentra los 5 títulos con las puntuaciones más bajas en IMDb.

```python
netflix_minus5 = netflix_filtrado.nsmallest(5, 'IMDB Score').filter(['IMDB Score', 'Title'])
netflix_minus5
```

#### Ejercicio 20: Filtra todos los títulos que tienen información completa en las columnas 'title', 'country' y 'IMDB Score'.

```python
filtro_null1 = netflix_filtrado['title'].notnull()
filtro_null2 = netflix_filtrado['country'].notnull()
filtro_null3 = netflix_filtrado['IMDB Score'].notnull()

not_nulos = netflix_filtrado['title'][filtro_null1 & filtro_null2 & filtro_null3]
not_nulos.head()
```
